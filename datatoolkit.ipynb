 The difference in npYarray, npYasarray and npYasanyarrayX
 The difference between Deep copy and shallow copyX
Ans Certainly! Here’s an explanation of the terms you mentioned, including the differences between `np.array`, `np.asarray`, and `np.asanyarray`, as well as the differences between deep copy and shallow copy.

### 1. Difference Between `np.array`, `np.asarray`, and `np.asanyarray`

- **`np.array()`**:
  - **Purpose**: Creates a new NumPy array from a given object (like a list, tuple, etc.).
  - **Behavior**: It always creates a new array regardless of the input type. This means that even if the input is already a NumPy array, `np.array()` will create a copy of that array.
  - **Example**:
    ```python
    import numpy as np

    arr1 = np.array([1, 2, 3])
    arr2 = np.array(arr1)  # Creates a copy
    ```

- **`np.asarray()`**:
  - **Purpose**: Converts an input to an array, but does not create a copy if the input is already an array.
  - **Behavior**: It will return the original array if the input is already a NumPy array, which makes it more memory-efficient in some cases.
  - **Example**:
    ```python
    arr1 = np.array([1, 2, 3])
    arr2 = np.asarray(arr1)  # Does not create a copy, returns the same array
    ```

- **`np.asanyarray()`**:
  - **Purpose**: Similar to `np.asarray()`, but it allows for subclasses of ndarray (like masked arrays) to be returned as arrays.
  - **Behavior**: It does not create a copy if the input is already an ndarray or any subclass of ndarray, but will return the input as it is.
  - **Example**:
    ```python
    arr1 = np.array([1, 2, 3])
    arr2 = np.asanyarray(arr1)  # Returns the same array, regardless of the subclass
    ```

### Summary:
- `np.array()`: Always creates a new array.
- `np.asarray()`: Converts to an array without copying if the input is already an array.
- `np.asanyarray()`: Like `np.asarray()`, but also accepts subclasses of ndarray.

---

### 2. Difference Between Deep Copy and Shallow Copy

- **Shallow Copy**:
  - **Definition**: A shallow copy creates a new object, but it does not create copies of nested objects. Instead, it copies references to the nested objects.
  - **Behavior**: Changes made to nested objects in the original will reflect in the shallow copy, as both share the same reference to the nested objects.
  - **Example**:
    ```python
    import copy

    original = [[1, 2, 3], [4, 5, 6]]
    shallow_copied = copy.copy(original)

    shallow_copied[0][0] = 99  # Modify the nested object
    print(original)  # Output: [[99, 2, 3], [4, 5, 6]]
    ```

- **Deep Copy**:
  - **Definition**: A deep copy creates a new object and recursively creates copies of all nested objects. This means that the new object is completely independent of the original.
  - **Behavior**: Changes made to nested objects in the original will not affect the deep copy.
  - **Example**:
    ```python
    import copy

    original = [[1, 2, 3], [4, 5, 6]]
    deep_copied = copy.deepcopy(original)

    deep_copied[0][0] = 99  # Modify the nested object
    print(original)  # Output: [[1, 2, 3], [4, 5, 6]]
    ```

### Summary:
- **Shallow Copy**: Copies the object but shares references to nested objects.
- **Deep Copy**: Creates a completely independent copy of the object and all nested objects.

Ans.To create a Seaborn scatter plot of two random arrays, color the points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot', you can use the following code snippet:

### Code Example

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create two random arrays
x_scatter = np.random.rand(100) * 10  # Random numbers between 0 and 10
y_scatter = np.random.rand(100) * 10  # Random numbers between 0 and 10

# Step 2: Create a DataFrame
data = pd.DataFrame({'X': x_scatter, 'Y': y_scatter})

# Step 3: Create a column for quadrant colors
def get_quadrant(row):
    if row['X'] >= 5 and row['Y'] >= 5:
        return 'Quadrant I (Red)'
    elif row['X'] < 5 and row['Y'] >= 5:
        return 'Quadrant II (Blue)'
    elif row['X'] < 5 and row['Y'] < 5:
        return 'Quadrant III (Green)'
    else:
        return 'Quadrant IV (Orange)'

data['Quadrant'] = data.apply(get_quadrant, axis=1)

# Step 4: Create a Seaborn scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='X', y='Y', hue='Quadrant', palette={'Quadrant I (Red)': 'red',
                                                                 'Quadrant II (Blue)': 'blue',
                                                                 'Quadrant III (Green)': 'green',
                                                                 'Quadrant IV (Orange)': 'orange'})

# Step 5: Customize the plot
plt.title('Quadrant-wise Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(title='Quadrants')
plt.grid(True)
plt.show()
```

### Explanation of the Code:
1. **Creating Random Arrays**:
   - Two random arrays, `x_scatter` and `y_scatter`, are generated with values between 0 and 10.

2. **Creating a DataFrame**:
   - A DataFrame named `data` is created containing the random arrays.

   - A new column named `Quadrant` is added to the DataFrame based on the values of `X` and `Y`. The function `get_quadrant` assigns a label for each quadrant.

4. **Creating the Scatter Plot**:
   - A Seaborn scatter plot is generated using `sns.scatterplot()`, with points colored according to their assigned quadrant. A custom color palette is defined.

### Code Example

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# Step 1: Prepare random categorical data
categories = ['A', 'B', 'C', 'D', 'E']  # Define categories
values = np.random.randint(10, 100, size=len(categories))  # Random values between 10 and 100

# Step 2: Create a DataFrame and a ColumnDataSource
data = pd.DataFrame({'Category': categories, 'Value': values})
source = ColumnDataSource(data=data)

# Step 3: Create a Bokeh figure
p = figure(x_range=categories, title='Random Categorical Bar Chart',
           x_axis_label='Categories', y_axis_label='Values',
           plot_height=400, toolbar_location=None)

# Step 4: Add color to bars based on values
color_mapping = ['red' if value < 30 else 'yellow' if value < 60 else 'green' for value in values]

# Step 5: Create bars
p.vbar(x='Category', top='Value', source=source, width=0.9, color=color_mapping, legend_field='Category')

# Step 6: Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

# Step 7: Customize the plot
p.legend.title = "Categories"
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.y_range.end = max(values) + 10

# Step 8: Show the plot
output_notebook()  # Use this line to display in Jupyter Notebook. Remove if running in a script.
show(p)
```

### Code Example

```python
import plotly.graph_objects as go
import numpy as np

# Step 1: Generate random data
x = np.linspace(0, 10, 100)  # x values from 0 to 10
y = np.random.rand(100) * 10  # y values as random numbers between 0 and 10

# Step 2: Create a line plot
fig = go.Figure()

# Step 3: Add a line trace
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data', line=dict(color='blue')))

# Step 4: Label the axes and set the title
fig.update_layout(title='Simple Line Plot',
                  xaxis_title='X-axis',
                  yaxis_title='Y-axis')

# Step 5: Show the plot
fig.show()
```

   - `y` values are generated as random numbers multiplied by 10.

2. **Create a Line Plot**:
   - A Plotly figure is created with `go.Figure()`.

3. **Add a Line Trace**:
   - A line trace is added to the figure using `go.Scatter()`, with specified x and y data. The `mode` is set to `'lines'` to create a line plot, and the line color is specified.

4. **Label the Axes and Set the Title**:
   - The x-axis and y-axis are labeled, and the plot title is set using `update_layout()`.

Q21.Usin Plotly, create an interactive pie chart of randomly enerated data, add labels and percentaes, set
the title as 'Interactive Pie Chart'.
Ans To create an interactive pie chart of randomly generated data using Plotly, you can use the following code. This example generates random data for the pie chart and adds labels and percentages, with the title set as 'Interactive Pie Chart'.

### Code Example

```python
import plotly.graph_objects as go
import numpy as np

# Step 1: Generate random data
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']  # Define labels
values = np.random.randint(10, 100, size=len(labels))  # Random values between 10 and 100

# Step 2: Create a pie chart
fig = go.Figure(data=[go.Pie(labels=labels, values=values,
                               textinfo='label+percent',  # Show labels and percentages
                               hole=0.3)])  # Optional: create a donut chart by setting hole size

# Step 3: Set the title
fig.update_layout(title='Interactive Pie Chart')

# Step 4: Show the plot
fig.show()
```

### Explanation of the Code:
1. **Generate Random Data**:
   - A list of labels representing different categories is defined.
   - Random values are generated for these categories using `np.random.randint()` to create integers between 10 and 100.

2. **Create a Pie Chart**:
   - A Plotly figure is created using `go.Figure()`, and a pie chart is added with `go.Pie()`. The `textinfo` parameter is set to `'label+percent'` to display both labels and percentages on the pie slices. The `hole` parameter is set to 0.3 to create a donut chart; you can remove this parameter if you want a standard pie chart.

3. **Set the Title**:
   - The title of the chart is set using `update_layout()`.

4. **Show the Plot**:
   - The pie chart is displayed using `fig.show()`.
